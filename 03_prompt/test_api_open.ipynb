```python
# load the data
import json

with open("test_emails.json", "r") as file:
    emails = json.load(file)


emails
```

```python
def classify_email(prompt):
    response = client.chat.completions.create(
            model="gpt-4o",
            messages=[
                {
                    "role": "user",
                    "content": prompt
                }
            ],
            temperature=0 # experiemntation : handson - do some experiemtn here and notedown your finding
        )
    return response.choices[0].message.content


```

```python
classify_email('write 100 word short email to a customer thanking them for their purchase and asking for feedback')
```
